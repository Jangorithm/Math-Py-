```python
age = [31,27,25,24,21,44,36,29,35,28,16,26,21,37,26,22,30,29,40,33,36,26,22,42,25,28,39,30,27,31]

age15 = []
age20 = []
age25 = []
age30 = []
age35 = []
age40 = []

def age_categories(age):
    if 15 <= age < 20:
        age15.append(age)
    elif 20 <= age < 25:
        age20.append(age)
    elif 25 <= age < 30:
        age25.append(age)
    elif 30 <= age < 35:
        age30.append(age)
    elif 35 <= age < 40:
        age35.append(age)
    else:
        age40.append(age)        

for i in range(len(age)):
    age_categories(age[i])
```

```python

import matplotlib.pyplot as plt

data = [len(age15), len(age20), len(age25), len(age30), len(age35), len(age40)]
labels = ['age15', 'age20', 'age25', 'age30', 'age35', 'age40']

plt.figure(figsize=(8,8))
plt.title("Pie Chart")
plt.pie(data, autopct='%.1f%%')
plt.legend(labels)
plt.show()
```

```python
#확률적경사하강법
# 실행할 때마다 값이 달라짐. 보통 74.xx 가 나오면 정답

from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# 학습률에 따라 값이 달라짐
# 과정이 맞으면 정답으로 채점함

from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib as plt

X_train = np.array([10,22,30,38,50])
y_train = np.array([41,45,62,75,85])

X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)

scalerX = StandardScaler()
scalerX.fit(X_train)
X_train = scalerX.transform(X_train)

scalerY = StandardScaler()
scalerY.fit(y_train)
y_train = scalerY.transform(y_train)

model = SGDRegressor(verbose=1)
model.fit(X_train, y_train)

X_test = np.array([45]).reshape(-1,1)
X_test = scalerX.transform(X_test)

y_pred = model.predict(X_test).reshape(1,-1)
y_pred_inverse = scalerY.inverse_transform(y_pred)
print("X값이 45일 때 출력값: ", y_pred_inverse)
```

<h3>[몬티 홀 문제(Monti Hall Problem]</h3>
    
세 개의 문 중에 하나를 선택하여 문 뒤에 있는 선물을 가지는 게임쇼에 참가했다. 한 문 뒤에는 자동차가 있고, 나머지 두 문 뒤에는 염소가 있다. 이 때, 어떤 사람이 1번 문을 선택했을 때, 게임쇼 진행자는 3번 문을 열어 문 뒤에 염소가 있음을 보여주면서 1번 대신 2번을 선택하겠냐고 물었다. 참가자가 자동차를 가지려할 때, 원래 선택했던 번호를 바꾸는 것이 유리할까?

```python


import random

door = ['door1', 'door2', 'door3']
no_change = 0
change = 0


for i in range(100):
    car = random.randint(0,2)
    player_choice = random.randint(0,2)
    
    empty_door = []
    
    for j in range(3):
        if j != player_choice and j != car:
            empty_door.append(door[j])
            
    com = random.sample(empty_door, 1)[0]
    
    
    if player_choice == car:
        no_change += 1
        
    empty_door.remove(com)
    
    if not empty_door:
        change += 1
        
        
print("기존 선택 유지하기: ", no_change,"%")
print("번호 바꾸기:", change, "%")
```
