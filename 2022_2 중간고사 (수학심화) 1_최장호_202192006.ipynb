### 문제4 다음 방정의 해를 구하시오

$$
3x_1 + 0x_2 + 2x_3 = 7
$$
$$
2x_1 + 0x_2 - 2x_3 = 10
$$
$$
0x_1 + x_2  +  x_3 = 5  
$$


```python
from sympy import *
import numpy as np

x1,x2, x3 = symbols('x1 x2 x3')

eq1 = 3*x1 + 0*x2 + 2*x3 - 7
eq2 = 2*x1 + 0*x2 - 2*x3 - 10
eq3 = 0*x1 + x2 + x3 - 5

sol = solve((eq1, eq2, eq3), dict=True)
sol
```

```python
import numpy as np
a = np.array([[3,0,2],[2,0,-2],[0,1,1]])
b = np.array([7,10,5])

np.linalg.solve(a,b)
```

### 문제 6 다음 행렬의 고유값과 고유벡터 그리고 고유분해(eigendecpmposition)를 구하세요.

$$
A=
\begin{bmatrix}
     2 & 3\\
     9 & 5
\end{bmatrix}
$$

- 1) 고유값
- 2) 고유벡터
- 3) 고유분해

```python
A = np.array([[2,3],[9,5]])

w2, v2 = np.linalg.eig(A)
w2, v2
```

```python
v2_inv = np.linalg.inv(v2)
v2_inv
```

```python
v2 @ np.diag(w2) @ v2_inv
```

###  문제8 다음 함수를 미분하고 그래프를 그리세요

$$
y = \frac{1}{1+e^{(-x)}}
$$

# Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, sigmoid(xx))

plt.xlabel("$x$")
plt.ylabel("$\sigma(x)$")
plt.show()
```

# sigmoid 미분 그래프

```python
def D_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

xx = np.linspace(-10, 10, 100)
plt.plot(xx, D_sigmoid(xx))

plt.show()
```

### (보너스 문제) 임의의 본인 사진을 가지고 특이값분해를 하여 특이값이 커비면서 해상도 변화를 보는 프로그램하기

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

plt.style.use('classic')
img = Image.open("C:/Users/kowm6/OneDrive/바탕 화면/시험에 사용할 사진.jpg")
plt.figure(figsize=(6, 4))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python

imggray = img.convert('LA')

imgmat = np.array(list(imggray.getdata(band=0)), float)

imgmat.shape = (imggray.size[1], imggray.size[0])

plt.figure(figsize=(6, 4))
plt.imshow(imgmat, cmap='gray')
plt.axis('off')
plt.show()
```

```python
U, D, V = np.linalg.svd(imgmat)
```

```python
imgmat.shape
```

```python
reconstimg = np.matrix(U[:, :2]) * np.diag(D[:2]) * np.matrix(V[:2, :])
plt.figure(figsize=(6, 4))
plt.imshow(reconstimg, cmap='gray')
plt.axis('off')
plt.show()
```

```python
for i in [5, 15, 35, 55, 100]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i, :])
    plt.figure(figsize=(6, 4))
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.axis('off')
    plt.show()
```
